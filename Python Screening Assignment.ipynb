## 1. Create a function in python to read the text file and replace specific content
of the file.

File name example.txt
Origin file content This is a placement assignment
Replace string Placement should be replaced by

screening.

Replaced file content This is a screening assignment

Creating original file 

```python
f=open('example.txt','w')
f.write('This is a placement assignment')
f.close()
```

Creating file with replaced content

```python
# Read in the file
with open('example.txt', 'r') as file :
  filedata = file.read()

# Replace the target string
filedata = filedata.replace('placement', 'screening')

# Write the file out again
with open('example.txt', 'w') as file:
  file.write(filedata)
```

2. Demonstrate use of abstract class, multiple inheritance and decorator in
python using examples.

```python
from abc import ABC, abstractmethod

class flight1(ABC):
    def run(self):
        print("Running 200km")

    @abstractmethod
    def fly(self):
        pass

class flight2(flight1):
    fname='Indigo'
    def __init__(self,fid):
        self.fid=fid
        print("flight id is ",self.fid)
        
    def fly(self):
        print("Flying")
        

class flight3(flight2):

    def ac(self):
        print("AC")

class flight4(flight3):
    def run(self):
        print("Running 400 km")
f1=flight4(1)
print(flight4.fname)
f1.run()
f1.fly()
f1.ac()
print("---------------------------")
f2=flight3(2)
f2.run()
        
    
```

# multiple inheritance in python

```python
class father:
    def bmw(self):
        print("father bmw")
class mother:
    def bmw(self):
        print("mother bmw")
class son(mother,father):
    def cycle(self):
        print("son cycle")
        
s=son()
s.cycle()
s.bmw()
```
